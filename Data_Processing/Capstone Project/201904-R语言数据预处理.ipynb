```python
import pandas as pd
data = pd.read_csv("E:\\Capstone Project\\Data\\Taxi_area\\pick201904attributes.csv")
```

```python
data
```

```python
data['periods'] = ''
for j in range(len(data)):
    data.loc[j,'periods'] = data.loc[j,'time'][11]
data
```

```python
for j in range(len(data)):
    if data.loc[j,'periods'] == '0' or data.loc[j,'periods'] == '1':
        data.drop([j],axis=0, inplace = True)
data
```

```python
data = data.reset_index(drop=True)
```

```python
data[data['time'] == '2019-04-22 2'].index
```

```python
train = data.iloc[0:27128,0:6]

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_date = le.fit(train['time'])
train['time'] = df_date.transform(train['time'])

train
```

```python
test = data.iloc[27128:,0:6]

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_date = le.fit(test['time'])
test['time'] = df_date.transform(test['time'])

test
```

```python
all = data.iloc[:,0:6]

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_date = le.fit(all['time'])
all['time'] = df_date.transform(all['time'])

all
```

```python
train.to_csv(r'E:\\Capstone Project\\Data\\Taxi_area\\pick201904train.csv',index=False,sep=',')
test.to_csv(r'E:\\Capstone Project\\Data\\Taxi_area\\pick201904test.csv',index=False,sep=',')
all.to_csv(r'E:\\Capstone Project\\Data\\Taxi_area\\pick201904all.csv',index=False,sep=',')
```
